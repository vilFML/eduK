### Ejercicio 1.5

Se le llama "Camel Case" a la convención de escribir una frase sin espacios, pero marcando el inicio de cada palabra poniendo su primera letra en mayúscula. Por ejemplo, la frase
```
"  Algoritmos y    estructuras de   datos   "
```
transformada a Camel Case queda así:
```
"AlgoritmosYEstructurasDeDatos"
```

Escriba una función que transforme a Camel Case y pruébela:

````python
def CamelCase(s):

    # usando diagrama de estados de forma iteracion
    # OUT: inicio i=0, sAux=''
    i = 0
    sAux = ''

    while (i < len(s)):
        # Sigue OUT si letra del string es ''
        if s[i] == ' ':
            # siguiente letra (indice sig)
            i += 1
        else:
            # si caracter no es espacio: entrada a IN
            # agrega letra mayuscula
            sAux = sAux + s[i].upper()
            # siguiente caracter
            i += 1

            #inicio bucle IN: se debe tener letra e i estar dentro del string
            while (s[i] != ' ' and i < len(s)):
                # si letra no es espacio, se mantiene dentro
                if s[i] != ' ':
                    # y se agrega letra a string
                    sAux = sAux + s[i]
                    i += 1
    
    return sAux
        
````

````python
print(CamelCase("    Algoritmos y    estructuras de   datos   "))
````
