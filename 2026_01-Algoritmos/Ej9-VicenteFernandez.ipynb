### Nodo2: 
busqueda binaria,
1. Si llave buscada coincide con llave almacenada: devolver almacenada
```py
if key == x:
    return key
```
2. Si buscada es menor que almacenada: devolver recursivamente a la izq
```py
if x < key:
    return self.izq.search(x)
```
3. Si buscada mayor que almacenada: devolver recursivamente a la der
```py
if x > key:
    return self.der.search(x)
```

### Nodo3: 
Almacena dos llaves y tiene 3 hijos:

hijo izq < key izq < hijo medio < key der < hijo der

luego, se revisan más casos.

1. Si buscada es igual a llave izq: devolver llave izq
```py
if x == keyIzq:
    return keyIzq
```
2. etc
```py
elif x == keyDer:
    return keyDer
elif x < keyIzq:
    return self.izq.search(x)
elif x > keyIzq and x < keyDer:
    return self.med.search(x)
elif x > keyDer:
    return self.der.search(x) 
```

hay que retornar puntero, entonces hay que retornar ```self```

````python
!pip install --upgrade aed-utilities

import aed_utilities as aed

class Nodo2:
    def __init__(self, izq, info, der):
        self.izq=izq
        self.info=info
        self.der=der

    def string(self):
      return "Nodo2("+str(self.info)+")"

    #Implementa esta función
    def search(self,x):
        key = self.info

        if key == x:
            return self
        elif x < key:
            return self.izq.search(x)
        elif x > key:
            return self.der.search(x)

class Nodo3:
    def __init__(self, izq, info1, med, info2, der):
        self.izq=izq
        self.info1=info1
        self.med=med
        self.info2=info2
        self.der=der

    def string(self):
      return "Nodo3("+str(self.info1)+","+str(self.info2)+")"

    #Implementa esta función
    def search(self,x):
        keyIzq = self.info1
        keyDer = self.info2
        
        if x == keyIzq:
            return self
        elif x == keyDer:
            return self
        elif x < keyIzq:
            return self.izq.search(x)
        elif x > keyIzq and x < keyDer:
            return self.med.search(x)
        elif x > keyDer:
            return self.der.search(x)

class Nodoe:
    def __init__(self):
        pass

    #Implementa esta función
    def search(self,x):
        return None 

    def __str__(self):
      return "☐"

class Arbol23:
    def __init__(self,raiz=Nodoe()):
        self.raiz=raiz

    def search(self,x):
        return self.raiz.search(x)

    def dibujar(self):
        drawer = aed.Tree23Drawer(Nodo2, Nodo3, Nodoe)
        drawer.draw_tree(self)
````
