Suponga que los campos ``info`` de los nodos externos de un árbol contienen solo números y escriba una función que pueda invocarse como ``formula.evaluar()``, que al ser ejecutada entregue el valor numérico  de la fórmula representada por el árbol. Modifique a continuación la definición de las clases ``Arbol``, ``Nodoi`` y ``Nodoe`` para que incluyan la nueva función ``evaluar()``. Note que tiene que implementar la función ``evaluar()`` en las tres clases.

```python
class Nodoi:
    def __init__(self, izq, info, der):
        self.izq=izq
        self.info=info
        self.der=der
    
    def evaluar(self):
        # en nodointerno funcion debiese retornar la evaluación de operacion en info con los datos en los nodos externos

        nroIzq = self.izq.evaluar()                                             # esto llama evaluar() en los nodos externos del interno
        nroDer = self.der.evaluar()
        operacion = self.info
        #DEBUG: print('Nodoizq: ', nroIzq,' operacion: ', operacion, ' NodoDer: ', nroDer)

        # se hace por casos según operacion q se tenga
        if (operacion == '+'):
            return (nroIzq + nroDer)
        elif (operacion == '-'):
            return (nroIzq - nroDer)
        elif (operacion == '*'):
            return (nroIzq * nroDer)
        elif (operacion == '/'):
            return (nroIzq // nroDer)
        else:
            print('Operación invalida.')

class Nodoe:
    def __init__(self, info=""):
        self.info=info

    def evaluar(self):
        # esta funcion solamente debiese ser invocada y solo retornar su número (el caso base)
        return self.info

class Arbol:
    def __init__(self,raiz=Nodoe()):
        self.raiz=raiz

    def evaluar(self):
        # esta funcion debiese invocar la fn evaluar desde la raiz
        return self.raiz.evaluar()
```

Pruébela a continuación:

```python
formula= Arbol(
            Nodoi(
                Nodoi(Nodoe(5),"+",Nodoe(2)),
                "*",
                Nodoi(
                    Nodoe(8),
                    "-",
                    Nodoi(Nodoe(9),"/",Nodoe(3))
                )
            )
        )   
print(formula.evaluar())
```
